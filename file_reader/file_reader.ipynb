```python
print('CLI FILE READER'.center(100, '_')+'\n')
path = str(input('Enter the address of the file to read : '))

try:
    with open(path, 'r', encoding = 'utf-8') as file:
        content = file.readlines()
        
        if content:
            for i in content:
                print(i.center(100))
                print('-'.center(100,'-'))

        else:
            print('File is empty!')

except FileNotFoundError:
    print('Invalid Address, File Not Found !')

except UnicodeDecodeError:
    print('Unknown File Format Error !!')

except Exception as e:
    print(f'Unknown Error {e}')
```
